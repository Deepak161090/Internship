```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
tags = soup.find_all(['h1','h2','h3','h4','h5','h6'])
```
