```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import selenium
from selenium import webdriver
import time
```

```python
driver = webdriver.Chrome(r"C:\chromedriver.exe")
driver.get('https://www.glassdoor.co.in/Salaries/index.htm')
driver.find_element_by_id('KeywordSearch').send_keys('Data Scientist')
driver.find_element_by_id('LocationSearch').clear()
driver.find_element_by_id('LocationSearch').send_keys('Noida')
driver.find_element_by_id('HeroSearchButton').click()
time.sleep(10)
```

```python
company_names=[]
salaries_no=[]
avg_sal=[]
min_sal=[]
max_sal=[]
for i in list(driver.find_elements_by_xpath("//p[@class='m-0']"))[:30]:
    company_names.append(i.text)
for i in list(driver.find_elements_by_xpath("//p[@class='css-1uyte9r css-1kuy7z7 m-0']"))[:10]:  
    salaries_no.append(i.text)
for i in list(driver.find_elements_by_xpath("//div[@class='col-2 d-none d-md-flex flex-row justify-content-end']"))[:10]:
    avg_sal.append(i.text.split('\n')[0])
for i in list(driver.find_elements_by_xpath("//div[@class='common__RangeBarStyle__values common__flex__justifySpaceBetween common__flex__container ']"))[:10]:
    min_sal.append(i.text.split('\n')[0])
    max_sal.append(i.text.split('\n')[1])
job_df=pd.DataFrame({'Company Name':company_names[1::3],
                         'Salary Number':salaries_no,
                         'Average Salary':avg_sal,
                         'Minimum Salary':min_sal,
                         'Maximum Salary':max_sal})
job_df.to_csv('Data Scientist GlassDoor Salary', index = False)
print(job_df)
```
