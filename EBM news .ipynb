```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import selenium
from selenium import webdriver
import time
```

```python
driver = webdriver.Chrome(r"C:\chromedriver.exe")
```

```python
date =[]
vertical=[]
author=[]
headline=[]
vertical_page=[]
description=[]
for i in range(162,260):
    driver.get("https://www.ebmnews.com/2018/page/%d" %i)
    for i in list(driver.find_elements_by_xpath("//h2[@class='title']")):
        headline.append(i.text)
    for i in list(driver.find_elements_by_xpath("//span[@class='time']")):
        date.append(i.text)
    for i in list(driver.find_elements_by_xpath("//i[@class='post-author author']")):
        author.append(i.text)
    for i in list(driver.find_elements_by_xpath("//div[@class='post-summary']")):
        description.append(i.text)
    for i in list(driver.find_elements_by_xpath("//h2[@class='title']/a")):
        vertical_page.append(i.get_attribute('href'))
```

```python
for i in vertical_page:
        driver.get(i)
        for i in list(driver.find_elements_by_xpath("//div[@class='post-meta-wrap clearfix']/div")):
            vertical.append(i.text)

            
news_df=pd.DataFrame({'Vertical':vertical,
                         'Headline':headline,
                         'Date':date,
                         'Description':description,
                         'Author':author})
news_df.to_csv('EBM news', index = False)
print(news_df)
```
