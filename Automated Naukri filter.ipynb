```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import selenium
from selenium import webdriver
import time
```

```python
driver = webdriver.Chrome(r"C:\chromedriver.exe")
driver.get('https://www.naukri.com/data-scientist-jobs?k=data%20scientist&cityType=25.9.31&ctcFilter=3to6')
job_names=[]
company=[]
location=[]
experience=[]
salary=[]
for i in list(driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"))[:10]:
    job_names.append(i.text)
for i in list(driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']"))[:10]:  
    company.append(i.text)
for i in list(driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']"))[:10]:
    location.append(i.text)
for i in list(driver.find_elements_by_xpath("//span[@class='ellipsis fleft fs12 lh16']"))[:30]:
    experience.append(i.text)
job_df=pd.DataFrame({'Job_Names':job_names,
                         'Company':company,
                         'location':location,
                         'Experience':experience[0::3],
                         'Salary':experience[1::3]})
job_df.to_csv('Data Scientist Naukri.com with Filters', index = False)
print(job_df)
```
