```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
```

```python
driver = webdriver.Chrome(r"C:\chromedriver.exe")
driver.get("https://www.imdb.com/india/top-rated-indian-movies/")
```

```python
Title=[]
Year=[]
Rating=[]
```

```python
title_list=driver.find_elements_by_xpath("//td[@class='titleColumn']")
```

```python
for i in title_list:
    name=i.text
    Title.append(name)
Title[0:10]
```

```python
ratings = driver.find_elements_by_xpath("//td[@class='ratingColumn imdbRating']")
for i in ratings:
    rate=i.text
    Rating.append(rate)
Rating[0:10]
```

```python
year=driver.find_elements_by_xpath("//span[@class='secondaryInfo']")
for i in year:
    y=i.text
    Year.append(y)
Year[:6]
```

```python
t=[]
for i in Title:
    m=i.split('.')[1]
    t.append(m)
t
```

```python
ti=[]
for i in t:
    n=i.split('(')[0]
    ti.append(n)
ti
```

```python
Movie_info=pd.DataFrame({})
Movie_info['Title'] = ti[:100]
Movie_info['Year']= Year[:100] 
Movie_info['Rating']=Rating[:100]
```
